```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
```

```python
import numpy as np
import pandas as pd
import random
rng = np.random.default_rng(1)
z = np.array(['Red', 'White', 'Blue', 'Blue', 'Red', 'Red'])
x = rng.integers(100,1000,10)
y = rng.choice([True, False], size =10, p=[0.7,0.3])
x, y, z
```

```python
data ={'Obs': [1,2,3,4,5,6],'X1':[0,2,0,0,-1,1], 'X2':[3,0,1,1,0,1], 'X3':[0,0,3,2,1,1], 'Y':['Red', 'Red', 'Red', 'Green', 'Green', 'Red']}
df = pd.DataFrame(data)
df = df.set_index('Obs')
```

```python
#Split dataset into training and testing sets
X = df[['X1','X2','X3']]
y = df['Y']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=50)
```

```python
#Create knn classifier
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
#Train model
knn.fit(X_train,y_train)
```

```python
#Make predictions  
y_pred = knn.predict(X_test)
y_pred
#print('Accuracy:', accuracy_score(y_test, y_pred))
#print('Classification Report:')
#print(classification_report(y_test, y_pred))
```

```python
Advertising= pd.read_csv('Advertising.csv')
Advertising
```

```python
#Split dataset into training and testing sets
X = Advertising[['TV', 'Radio', 'Newspaper']]
y = Advertising['Sales']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
#Create knn classifier
knn = KNeighborsRegressor(n_neighbors = 3)

#Train model
knn.fit(X_train,y_train)
```

```python
#Make predictions  
y_pred = knn.predict(X_test)
y_pred
```

```python
#Evaluate model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'Root mean square error: {rmse:.2f}')
```

```python
#Make new predition
milo = pd.DataFrame({'TV': [8900], 'Radio': [200], 'Newspaper':[500]})
Predict_milo_sale = knn.predict(milo)
print(f'Predicted Milo Sale: ${Predict_milo_sale[0]:.2f}')
```
